```python
data = [float(x) for x in accs_cifar.split()]
data = np.array(data).reshape(4,-1)


# Build the plot
fig, ax = plt.subplots(figsize=(6,6))
plt.rcParams.update({'font.size': 13})

ax.bar(np.arange(9), data.mean(axis=0) , yerr=data.std(axis=0), width=0.7, align='center', alpha=0.5, ecolor='black', capsize=10)
# ax.bar(np.array([0,1,2,3,4]) + .4, SVM_r.mean(axis=0) , yerr=SVM_r.std(axis=0), width=0.4, align='center', alpha=0.5, ecolor='black', capsize=15)
ax.set_ylabel('Acc')
# ax.set_xlabel('Num training data')
ax.set_xticks(np.arange(9))
ax.set_xticklabels(model_labels, rotation=90)
ax.set_title('Classification accuracy on CIFAR-10 dataset')
ax.yaxis.grid(True)
# plt.ylim(0,100)


# Save the figure and show
plt.tight_layout()
# plt.legend(['svm - linear kernel', 'svm - RBF kernel'])
# plt.savefig('res.png')

plt.plot(np.arange(9), [1/10]*9,linewidth=3)
plt.text(3, .115, r'chance level', fontsize=15)

plt.show()
```
